## Notebook 05 — Hypothesis Validation

**Objectives:** Statistically validate 3 flood risk hypotheses using chi-square and t-tests

**Inputs:** outputs/v1/featured_data.csv

**Outputs:** p-values and conclusions documented in README and displayed on dashboard Page 3

```python
# Notebook 05 — Hypothesis Validation
# Objectives: Statistically validate 3 flood risk hypotheses
# Inputs: outputs/v1/featured_data.csv
# Outputs: p-values and conclusions documented in README and dashboard
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, ttest_ind

df = pd.read_csv('../outputs/v1/featured_data.csv')
print(f"Shape: {df.shape}")
```

```python
# Hypothesis 1: Properties below 10m elevation have a higher flood risk
# Test: Chi-square test -> elevation_category vs flood_risk

contingency_table = pd.crosstab(df['elevation_category'], df['flood_risk'])
print("Contingency table:")
print(contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")

if p < 0.05:
    print("\nHypothesis 1 SUPPORTED — significant relationship between elevation and flood risk")
else:
    print("\nHypothesis 1 NOT SUPPORTED — no significant relationship found")
```

```python
# Hypothesis 2: Distance to river is strongest predictor of flood risk
# Test: t-test comparing flood risk for stations within 100m vs beyond

close = df[df['distance_to_river_m'] < 100]['flood_risk_encoded']
far = df[df['distance_to_river_m'] >= 100]['flood_risk_encoded']

t_stat, p = ttest_ind(close, far)
print(f"Close stations (within 100m): {len(close)}")
print(f"Far stations (beyond 100m): {len(far)}")
print(f"Mean risk close: {close.mean():.3f}")
print(f"Mean risk far: {far.mean():.3f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("\nHypothesis 2 SUPPORTED — stations closer to river have significantly higher flood risk")
else:
    print("\nHypothesis 2 NOT SUPPORTED — no significant difference found")
```

```python
# Hypothesis 3: Western counties have higher flood depth than Eastern counties
# Test: Independent t-test on region vs flood_depth_m

west = df[df['region'] == 'West']['flood_depth_m']
east = df[df['region'] == 'East']['flood_depth_m']

t_stat, p = ttest_ind(west, east)
print(f"West stations: {len(west)}")
print(f"East stations: {len(east)}")
print(f"Mean depth West: {west.mean():.3f}m")
print(f"Mean depth East: {east.mean():.3f}m")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("\nHypothesis 3 SUPPORTED — Western counties have significantly higher flood depth")
else:
    print("\nHypothesis 3 NOT SUPPORTED — no significant difference found")
```

```python
# Summary of hypothesis validation results
print("HYPOTHESIS VALIDATION SUMMARY")
print("=" * 50)
print("\nHypothesis 1: Elevation vs Flood Risk")
print("Test: Chi-square")
print("Result: SUPPORTED (p = 0.0000)")
print("Finding: Low elevation stations have significantly higher flood risk")

print("\nHypothesis 2: Distance to River vs Flood Risk")
print("Test: Independent t-test")
print("Result: SUPPORTED (p = 0.0000)")
print("Finding: Stations within 100m of river have mean risk 1.19 vs 0.74")

print("\nHypothesis 3: West vs East Flood Depth")
print("Test: Independent t-test")
print("Result: SUPPORTED (p = 0.0007)")
print("Finding: Western counties mean depth 1.028m vs Eastern 0.767m")
```
