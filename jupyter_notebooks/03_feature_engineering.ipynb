```python
# Notebook 03 — Feature Engineering
# Objectives: Transform cleaned data into ML-ready features
# Inputs: outputs/v1/cleaned_data.csv
# Outputs: outputs/v1/featured_data.csv
```

```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('../outputs/v1/cleaned_data.csv')
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
```

```python
# Create elevation category for Hypothesis 1
# Below 10m = high risk, 10-50m = medium, above 50m = low
def categorise_elevation(elevation):
    if elevation < 150:
        return 'Low'
    elif elevation < 220:
        return 'Medium'
    else:
        return 'High'

df['elevation_category'] = df['elevation_m'].apply(categorise_elevation)
print(df['elevation_category'].value_counts())
```

```python
# Create distance category for Hypothesis 2
def categorise_distance(distance):
    if distance < 100:
        return 'Very Close'
    elif distance < 200:
        return 'Close'
    else:
        return 'Far'

df['distance_category'] = df['distance_to_river_m'].apply(categorise_distance)
print(df['distance_category'].value_counts())
```

```python
# Encode flood risk as numeric for ML
risk_map = {'Low': 0, 'Medium': 1, 'High': 2}
df['flood_risk_encoded'] = df['flood_risk'].map(risk_map)
print(df['flood_risk_encoded'].value_counts())
```

```python
# Save featured data to outputs/v1/
df.to_csv('../outputs/v1/featured_data.csv', index=False)
print(f"Saved to outputs/v1/featured_data.csv")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
```
