```python
# Notebook 01 — Data Collection
# Objectives: Pull OPW water level station data from live API
# Inputs: OPW Real-Time Water Levels API — no key required
# Outputs: inputs/datasets/raw/ireland_flood_data.csv

import requests
import pandas as pd
import os

os.makedirs('../inputs/datasets/raw', exist_ok=True)
print("Directory ready")
```

```python
# Correct OPW GeoJSON endpoint — station locations across Ireland
url = "http://waterlevel.ie/geojson/"

response = requests.get(url)
print(f"Status code: {response.status_code}")
```

```python
# Parse the GeoJSON response
data = response.json()

# See the structure
print(f"Type: {type(data)}")
print(f"Keys: {data.keys()}")
print(f"Number of stations: {len(data['features'])}")
```

```python
records = []

for feature in data['features']:
    props = feature['properties']
    coords = feature['geometry']['coordinates']
    props['longitude'] = coords[0]
    props['latitude'] = coords[1]
    records.append(props)

df = pd.DataFrame(records)

print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head()
```

```python
# Save raw data — never modify this file directly
df.to_csv('../inputs/datasets/raw/ireland_flood_data.csv', index=False)
print(f"Saved successfully")
print(f"Shape: {df.shape}")
print(f"Location: inputs/datasets/raw/ireland_flood_data.csv")
```
